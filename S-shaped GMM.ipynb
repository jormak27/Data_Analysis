```python
import itertools

import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

from sklearn import mixture
from sklearn.externals.six.moves import xrange

# Number of samples per component
n_samples = 1500

# Generate random sample
np.random.seed(0)
X = np.zeros((n_samples, 2))
step = 4 * np.pi / n_samples

for i in xrange(X.shape[0]):
    x = i * step - 6
    X[i, 1] = x + np.random.normal(0, 0.2)
    X[i, 0] = x + 3 * (np.sin(x) + np.random.normal(0, .2))

with open('quizz.csv','w') as f:
    f.write('x,y\n')
    for x in X:
        f.write('%f,%f\n' % (x[0], x[1]))

plt.plot(X[:,0], X[:,1],'.')
```

```python
## Fit a GMM


#plt.figure(figsize=(8,6))

gmm = mixture.GMM(n_components=5, covariance_type='full', n_iter=100)
title = "Expectation-maximization"

gmm.fit(X)
print X.shape

color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm'])

splot = plt.subplot(1, 1, 1)
Y_ = gmm.predict(X)
for i, (mean, covar, color) in enumerate(zip(gmm.means_, gmm._get_covars(), color_iter)):
    v, w = linalg.eigh(covar)
    u = w[0] / linalg.norm(w[0])
        
    plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=color)

    # Plot an ellipse to show the Gaussian component
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    ell = mpl.patches.Ellipse(mean, v[0], v[1], 180 + angle, color=color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.5)
    splot.add_artist(ell)

plt.title(title)
```

```python
x = -3
Xpred = np.vstack((x*np.ones(100), np.arange(-8,8, 16.0/100))).T

pred = np.exp(gmm.score(Xpred))
print pred.shape

#plt.figure(figsize=(8,6))
for i, (w, color) in enumerate(zip(gmm.weights_,color_iter)):
    plt.plot(Xpred[:,1], pred[:], color = color)
```
