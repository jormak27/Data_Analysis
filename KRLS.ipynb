```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import spatial
    
from sklearn.svm import SVR

import krls

def kernel_function(X, Z):
    
    # kernel bandwidth
    sigma = 0.5
    # make sure X is a NumPy matrix        
    if not isinstance( X, np.matrix ):
        X = np.matrix( X )
    # make sure Z is a NumPy matrix        

    if not isinstance( Z, np.matrix ):
        Z = np.matrix( Z )

    n, m = X.shape[ 0 ], Z.shape[ 0 ]

    XX = np.multiply(X,X);  XX = XX.sum(axis=1)  
    ZZ = np.multiply(Z,Z);  ZZ = ZZ.sum(axis=1)
    d = np.tile(XX, (1, m)) + np.tile(ZZ.T, (n, 1)) - 2 * X * Z.T
    Kexpd = np.exp(-d.T/(2*sigma*sigma))             

    return np.array(Kexpd)


%matplotlib inline

np.random.seed(0)

# true function
Xgrid = np.linspace(0, 10, 500)[:, np.newaxis]
Ftrue = np.sin(Xgrid).ravel()

# Training data (with noise in Y)
X = np.sort(10 * np.random.rand(40, 1), axis=0)
Y = np.sin(X).ravel()

Y += 0.1*np.random.randn(len(X))

# plot data
plt.plot(Xgrid, Ftrue, '-', c='g', linewidth=3, alpha = 0.25)
plt.plot(X, Y, '.', c='b')
```

```python
###############################################################################
# Implement kernel regression model

class KRR: 
    
    def __init__(self, X, Y, h=0.5, sig=0.1):
        self.h = h
        self.sig = sig
        self.X = X
        self.Y = Y
        self.K = self.GKernel(X,X)      
       
        self.alpha = np.dot(np.linalg.inv(self.K+sig*sig*np.eye(len(Y))), Y)
        
    def predict(self, Xpred):
        Kpred = self.GKernel(Xpred,self.X)
        return np.dot(Kpred, self.alpha)
    
    def GKernel(self, X1,X2):
        d = spatial.distance_matrix(X1,X2)
        K = np.exp(-(d**2) / (2*self.h*self.h))
        return K
```

```python
# Initialize a KRR instance and make predictions

h = 1
eps = 0.2

krr = KRR(X,Y,h,0.1)
Ypred = krr.predict(Xgrid)

svr_rbf = SVR(kernel='rbf', C=1e1, gamma=(1/(2*h*h)), epsilon = eps).fit(X,Y)
Ypred_svr = svr_rbf.predict(Xgrid)
```

```python
ald_thresh = 0.2   # linear dependence threshold, 0 means every sample is novel
maxsize = 50       # dictionary size in samples

adaptive = True      # data-adaptive dictionary clean up when maxsize reached
forget_rate = 0.0     # force to eliminate the oldest entries from time to time 

kp = krls.krls(kernel_function, ald_thresh, X[0], Y[0], maxsize, adaptive, forget_rate)
for i in range(len(Y)-1): 
  kp.update(X[i+1], Y[i+1])

Ypred_krls = kp.query(Xgrid)    
    
plt.plot(Xgrid, Ftrue, '-', c='g', linewidth=3, alpha = 0.25)   # plot true
plt.plot(Xgrid, Ypred, '-k')  # plot KRR

plt.plot(X, Y, '.', c='b')  # plot data 

plt.plot(Xgrid, Ypred_krls, '-r') # plot KRLS
plt.plot(kp.dp.Dict, kp.dp.Targ,'o') #plot dict entries

plt.show()
```
